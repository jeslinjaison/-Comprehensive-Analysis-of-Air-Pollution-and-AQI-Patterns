</h2>
The Air Quality Data in India dataset is a comprehensive environmental dataset that contains air pollution measurements collected from various cities and monitoring stations across India. The data was gathered from the Central Pollution Control Board (CPCB) and includes important pollutants such as PM2.5, PM10, NO₂, SO₂, CO, O₃, and overall Air Quality Index (AQI) values. This dataset helps in understanding the air pollution levels and environmental conditions in different regions of India over time.

```python
# Missing value is very high in Xylene column so drop that column
df.drop('Xylene', axis=1, inplace= True)
```

```python
# fill missing values using median
median_col=['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene']
df[median_col]= df[median_col].fillna(df[median_col].median())
```

```python
df[median_col].isna().sum()
```

```python
# fill missing values using mean
df['AQI'] = df['AQI'].fillna(df['AQI'].mean())
df['AQI']
```

```python
def aqi_category(aqi):
    if aqi <= 50:
        return "Good"
    elif aqi <= 100:
        return "Satisfactory"
    elif aqi <= 200:
        return "Moderate"
    elif aqi <= 300:
        return "Poor"
    elif aqi <= 400:
        return "Very Poor"
    else:
        return "Severe"

df['AQI_Bucket'] = df['AQI_Bucket'].fillna(
    df['AQI'].apply(aqi_category)
)
df['AQI_Bucket']
```

    font-size:30px;
    font-weight:bold;">
    ●  Detect and Remove outliers using IQR  
</h2>  

```python
Outlier_cols = ['AQI', 'PM2.5', 'PM10', 'NO2', 'SO2', 'CO']

for col in Outlier_cols:
    
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    
    IQR = Q3 - Q1
    print(IQR)
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    print(lower_limit,upper_limit)
    df = df[(df[col] >= lower_limit) & (df[col] <= upper_limit)]
    
```

```python
#add season column based on date
def get_season(date):
    month = date.month
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

df['Season'] = df['Date'].apply(get_season)
```

```python
df['pollution_Status']= np.where(df['AQI']>200,'High','Low')
```
